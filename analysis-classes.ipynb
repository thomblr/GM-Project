# Analysis of dataset metadata_2013

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import community as cl
import networkx as nx 
```

```python
df = pd.read_table('data/metadata_2013.txt', names=['i', 'class', 'gender'])
df.head()
```

```python
classes = df.groupby('class')
classes.count()
```

```python
labels = df['class'].unique()
values = df.groupby('class')['i'].count()
print(values)
plt.pie(values, labels=labels, autopct='%1.1f%%')
```

```python
genders = df.groupby('gender')
genders.describe()
```

```python
labels = df.gender.unique()
values = df.groupby('gender')['i'].count()
plt.pie(values, labels=labels, autopct='%1.1f%%')
```
