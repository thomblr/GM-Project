# Analysis of the dataset High-School_data_2013

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import networkx as nx 
from datetime import datetime
```

```python
highschool = pd.read_csv('data/High-School_data_2013.csv', sep=" ", names=['t', 'i', 'j', 'Ci', 'Cj'])
highschool.describe()
```

```python
# datetime.utcfromtimestamp(1385989980).strftime('%Y-%m-%d %H:%M:%S')
dates = list(pd.unique(highschool['t']))
new_dates = []
for date in dates:
    new_dates.append(datetime.utcfromtimestamp(date).strftime('%Y-%m-%d'))

unique_dates = list(pd.unique(new_dates))

highschool['day'] = highschool.apply(lambda row: datetime.utcfromtimestamp(row.t).strftime('%d'), axis=1)
count = highschool.groupby('day').count()

count['t'].plot.bar()
plt.show()
```
